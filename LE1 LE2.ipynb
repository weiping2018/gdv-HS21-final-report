```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
from fitter import Fitter, get_common_distributions, get_distributions 
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import csv
df_1 = pd.read_csv('out_1.csv', index_col=False)
df_2 = pd.read_csv('out_2.csv', index_col=False)
```

```python
## remove the NAN columns and rows
## unify/rename the feature names of the two sets of data
df_1=df_1.loc[:67,'Datum':'Geschwindigkeit [m/s]'].rename({'Datum': 'Date', 'Masse [kg]': 'm [kg]','Geschwindigkeit [m/s]':'v [m/s]'}, axis='columns')
df_2=df_2.loc[:31,'Date':'v [m/s]']
print(df_1.head())
print(df_2.head())
```

```python
##  add new columns to identify the dataset groups
##  group two dataframe as one
df_1['dataset'] = 'Zone 1'
df_2['dataset'] = 'Zone 2'
df_100 = df_1.append(df_2)
```

# Transformation

New features generation

```python
##  generate the new feature Aufprallenergie/Kinetic energy[kJ], based one the Kinetic energy (KE) calculation formula:KE = 0.5 * mv2  (J;kg;m/s)
df_100['KE [kJ]']=(0.5/1000)* df_100['m [kg]']*((df_100['v [m/s]'])**2)
## generate new feature Time by merging 'Date' and 'Uhrzeit'
df_100['Time']=pd.to_datetime(df_100['Date'] + ' ' + df_100['Uhrzeit'])
df_100 = df_100.loc[:,'m [kg]':'Time']
df_100.sort_values("Time", inplace=True)
df_100
```

```python
## generate new feature  Time diff -1 [h]: Time difference to the previous event
df_100['Time diff -1 [h]'] = (df_100['Time'] -df_100['Time'].shift(1)).astype('timedelta64[h]')
df_100['Time diff -1 [h]'].iloc[0] = 0
```
